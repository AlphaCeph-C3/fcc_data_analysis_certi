```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
df = pd.read_csv("./epa-sea-level.csv", index_col="Year")
df.tail()
```

```python
plt.figure(figsize=(14,9))
plt.scatter(df.index, df["CSIRO Adjusted Sea Level"])
res = linregress(df.index, df["CSIRO Adjusted Sea Level"])
line1_x_value = np.arange(1880, 2051)
line1_y_value = res.slope*line1_x_value + res.intercept
recent_years_df = df[df.index >=2000]
line2_res = linregress(recent_years_df.index, recent_years_df["CSIRO Adjusted Sea Level"])
line2_x_value = np.arange(2000, 2051)
line2_y_value = line2_res.slope*line2_x_value + line2_res.intercept
plt.plot(line1_x_value,line1_y_value, "-.", color='black', label="since 1880 data")
plt.plot(line2_x_value, line2_y_value, "--", color="red", label="since 2000 data")
plt.xlabel("Years")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
```
