```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("medical_examination.csv")
df.head()
```

```python
# df["overweight"] = ((df["weight"] * 10000) / (df["height"] ** 2) > 25) * 1
df["overweight"] = ((df["weight"] * 10000) / (df["height"] ** 2)).apply(lambda x: 1 if x > 25 else 0)
df
```

```python
df["cholesterol"] = np.where(df["cholesterol"] == 1, 0, 1)
df["gluc"] = np.where(df["gluc"] == 1, 0, 1)
dfm = pd.melt(df,id_vars=["cardio"], value_vars=["active", "alco", "cholesterol", "gluc", "overweight", "smoke"])
dfm
```

```python
g = sns.catplot(x="variable", data =dfm, col="cardio", kind="count", hue="value")
g.set_ylabels("total")
fig= g.fig
# fig.savefig("catplot.png")
```

```python
axis = fig.axes[0]
print(axis.get_xlabel())
print(axis.get_ylabel())
print(axis.get_xaxis().get_majorticklabels())
```

```python
mask = (df["ap_lo"] <= df["ap_hi"]) & (df["height"] >= df["height"].quantile(0.025)) & (df["height"] <= df["height"].quantile(0.975)) & (df["weight"] >= df["weight"].quantile(0.025)) & (df["weight"] <= df["weight"].quantile(0.975))
df_heat = df[mask]
df_heat
```

```python
corr = df_heat.corr()
corr
```

```python
corr_mask = np.triu(corr)
fig, ax = plt.subplots(figsize=(10, 9))
sns.heatmap(corr, fmt=".1f", annot=True, linewidths=0.1,square=True, mask=corr_mask, cbar_kws={"shrink": 0.7}, vmin=-0.08, vmax=0.24, ax=ax)
fig.savefig("heatmap.png")
```
